```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Plot Distribution of Scores attained by a group

```python
scores = pd.DataFrame(np.array([
    7,
    3,
    6,
    2,
    5,
    10,
    5.5,
    2.5,
    7.5,
    6.5,
    4.5]), columns=['scores'])
```

```python
from matplotlib import ticker

sns.set_style("darkgrid")
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})

sns.histplot(scores, kde=True, kde_kws=dict(cut=3), alpha=.4, ax=ax_hist)
sns.rugplot(scores, x='scores', hue='scores', palette='husl', ax=ax_hist)

sns.boxplot(scores, orient='horizontal', ax=ax_box); 
sns.despine(ax=ax_box, left=True); ax_box.set(yticks=[])

ax_box.text(scores.mean()[0], -.1, f'{scores.mean()[0]:.1f}', ha='center', size='small',color='w', weight='semibold')
ax_box.text(scores.max()[0], -.1, f'{scores.max()[0]:.0f}', ha='left', size='small',color='red', weight='semibold')
ax_box.text(scores.min()[0], -.1, f'{scores.min()[0]:.0f}', ha='right', size='small',color='red', weight='semibold')
ax_box.text(scores.quantile(q=.25)[0], -.1, f'{scores.quantile(q=.25)[0]:.0f}', ha='right', size='small',color='g', weight='semibold')
ax_box.text(scores.quantile(q=.75)[0], -.1, f'{scores.quantile(q=.75)[0]:.0f}', ha='left', size='small',color='g', weight='semibold')

ax_hist.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.1f}"))

plt.legend([])
plt.xlim([0,15])
plt.title('Quiz 1 (ROC) - Scores distribution')
```
